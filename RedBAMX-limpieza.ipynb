```python
import os
import urllib.request
import datetime
import pandas as pd
import numpy as np
```

```python
# DataFrame vacío para almacenar los datos
data_combined_2003_20 = pd.DataFrame()
data_combined_2021_22 = pd.DataFrame()

# Descarga y concatena los datos para cada año
for year in range(2003, 2021):
    url = f"http://infosiap.siap.gob.mx/gobmx/datosAbiertos/ProduccionAgricola/Cierre_agricola_mun_{year}.csv"
    data1 = pd.read_csv(url, encoding='latin-1')
    data_combined_2003_20 = pd.concat([data_combined_2003_20, data1])

for year in range(2021, 2023):
    url = f"http://infosiap.siap.gob.mx/gobmx/datosAbiertos/ProduccionAgricola/Cierre_agr_mun_{year}.csv"
    data2 = pd.read_csv(url, encoding='latin-1')
    data_combined_2021_22 = pd.concat([data_combined_2021_22, data2])

# Reinicia el índice del DataFrame combinado
data_combined = pd.concat([data_combined_2003_20, data_combined_2021_22])
data_combined.reset_index(drop=True, inplace=True)
```

```python
data_combined.to_csv('data_combined_2003_2023.csv', index=False)
```
