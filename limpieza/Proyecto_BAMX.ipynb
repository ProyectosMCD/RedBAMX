# Combinación de dataframes

## Merge con Concat
Se descarto debido a que al realizar la operación daba valores NaN

```python
dataframes = [df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022]
df_merged = pd.concat(dataframes, axis=0, ignore_index=True)

# Mostrar información del DataFrame resultante
print(df_merged.head())  # Muestra las primeras filas del DataFrame combinado
```

## Merge
Se uso merge para fusionar los dataframes, se realizo la fusión de manera exitosa

```python
df_merge = df2022.merge(df2021, on="Cultivo", how="right")
df_merge1 = df_merge.merge(df2020, on="Cultivo", how="right")
df_merge2 = df_merge1.merge(df2019, on="Cultivo", how="right")
df_merge3 = df_merge2.merge(df2018, on="Cultivo", how="right")
df_merge4 = df_merge3.merge(df2017, on="Cultivo", how="right")
df_merge5 = df_merge4.merge(df2016, on="Cultivo", how="right")
df = df_merge6 = df_merge5.merge(df2016, on="Cultivo", how="right")
```
